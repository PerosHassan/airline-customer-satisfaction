# Logistic Regression - Airline Customer Satisfaction Prediction
### 3MTT NextGen Cohort - Step 17 Evaluation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# 1. Load the dataset dynamically
import os
csv_file = [f for f in os.listdir('.') if f.endswith('.csv')][0]
df = pd.read_csv(csv_file)
df = df.dropna()

# 2. Clean column names and map target variable
df.columns = df.columns.str.get_letter_num_only().str.lower() if hasattr(df.columns, 'str') else [c.lower().strip() for c in df.columns]
target_col = [c for c in df.columns if 'satisfaction' in c][0]

df[target_col] = df[target_col].astype(str).str.lower().map({
    'satisfied': 1, 'satisfaction': 1, '1': 1, '1.0': 1,
    'dissatisfied': 0, 'neutral or dissatisfied': 0, '0': 0, '0.0': 0
}).fillna(0)

# 3. Feature Engineering (One-Hot Encoding Categorical Variables)
X = df.drop(columns=[target_col])
X = pd.get_dummies(X, drop_first=True)
y = df[target_col]

print("Features shape:", X.shape)
print("Target distribution:\n", y.value_counts())
```

### Model Training and Validation Split
Splitting data into 70% training and 30% testing subsets ensures reliable classification testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Binomial Logistic Regression model
model = LogisticRegression(max_iter=1000, solver='liblinear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Classifier successfully trained on training partition.")
```

### Performance Diagnostics & Matrices
Generating precision, recall scores, and confusion matrix layouts to calculate predictive accuracies.

```python
cm = confusion_matrix(y_test, y_pred)
print("--- Confusion Matrix ---")
print(cm)

print("\n--- Classification Metrics ---")
print(classification_report(y_test, y_pred))

# Heatmap layout setup
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples')
plt.title('Passenger Satisfaction Confusion Matrix')
plt.show()
```
